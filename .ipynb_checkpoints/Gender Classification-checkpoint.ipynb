## Clasificadores binarios para la data de género

* Se desea entrenar un clasificador en la tarea de clasificar el género de un rostro de forma binaria (Hombre o mujer).
* La data de entrenamiento corresponde a 140 rostros de hombres y 140 rostros de mujeres que fueron preprocesados anteriormente de la siguiente manera:
    1. Haarcascade para identificar la ubicación del rostro.
    2. Data Agumentation -> Girar el rostro, rotar el rostro en -30, -15, 15 y 30 grados.
    3. Cortar el rostro y dejarlo en un tamaño de 100x100
* Para la data de entrenamiento se tienen 6 imagenes a partir de cada imagen. Por lo tanto la dimensión de la data de entrenamiento es de 840 imagenes de rostros de hombres y 140 rostros de mujeres.
* La data de teste corresponde a 60 imagenes de rostros hombre y 60 imagenes de rostros de mujeres.

### Extracción de características
* Para extraer las caracteristicas de las imágenes se va a utilizar el método LBP (Local Binary Pattern) que corresponde a un descriptor de textura de la imágen.
* Se van a utilizar 2 variaciones de LBP:
    1. LBP uniforme (nri): La cual nos entrega un histograma de 59 variables descriptoras.
    2. LBP uniforme: La cual nos entrega un histograma de 10 variables descriptoras.
    
### Entrenamiento
* Primero se va a optimizar los hiperparametros de cada algoritmo a probar.
* El entrenamiento se va a realizar por medio de Cross Validation (10 Folds)
* Luego se va a testear en la data de testeo.

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.metrics import *
from pyClass import ClasificadorBinario
import os

import pickle
```

```python
def setear_parametros(clf, params):
    for k, v in params.items():
        clf = clf.set_params(**{k: v})
    return clf

def pipeline_train(clf, X,y, n_pliegues=7, plot=0):
    ## Seleccionar mejores parametros
    clf_train = ClasificadorBinario.optimizar_parametros(clf[1], clf[2], X, y)
    pc = ClasificadorBinario.optimizar_punto_corte(clf_train,X, y, plot=0)
    
    ## Setear parametros en el clasificador
    clf_parametros = setear_parametros(clf[1], clf_train.best_params_)
    ## Entrenar clasificador con los parametros
    clf_parametros = ClasificadorBinario.entrenamiento(clf_parametros, X, y, n_pliegues=n_pliegues, pc=pc)
    
    return clf_parametros, clf_train.best_params_

def Testear(clf_parametros, X_test, y_test, plot=0):
    # Optimizar punto de corte del clasificador
    pc = ClasificadorBinario.optimizar_punto_corte(clf_parametros, X_test, y_test, plot=0)
    y_hat, pred = ClasificadorBinario.probar_clasificador(clf_parametros, X_test, y_test, pc=pc)
    p = np.where(pred>pc, 1, 0)
    return y_hat, p

def save_model(clf, filename):
    pickle.dump(clf, open(filename, 'wb'))
    
path = './DataFeatures'
os.listdir(path)
```

### LBP nri uniform

```python
dt = pd.read_csv(path+'/FeaturesData_nri_uniform_Train.csv')
dt_test = pd.read_csv(path+'/FeaturesData_nri_uniform_Test.csv')
print(dt.shape, dt_test.shape)

X = dt.drop('y', axis=1)
dt['y_num'] = np.where(dt['y']=='Male', 1, 0)
y = dt.loc[:,['y_num']]

X_test = dt_test.drop('y', axis=1)
dt_test['y_num'] = np.where(dt_test['y']=='Male', 1, 0)
y_test = dt_test.loc[:,['y_num']]
```

```python
clfs = ClasificadorBinario.clasificadores_prueba(random_state=123)
clfs_optimizados = []

for clf in clfs:
    print(clf[0])
    clf_parametros, params =  pipeline_train(clf , X,y, n_pliegues=7, plot=0)    
    clfs_optimizados.append((clf[0], clf_parametros, params))    ## Guardar todo en la lista
    y_hat, p = Testear(clf_parametros, X_test, y_test)
    filename = './Modelos/nri_uniform/{}_{}.sav'.format(clf[0], int(accuracy_score(y_hat, p)*100))
    save_model(clf_parametros, filename)
    print('-'*100)
```
